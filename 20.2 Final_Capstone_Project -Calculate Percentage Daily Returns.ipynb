# FINAL CAPSTONE PROJECT OVERVIEW

Before we get started, we need to define the following terms:
Open: Price at which a security first trades when stock market opens. ​

High: Highest price at which a given stock traded during a regular trading session. ​

Low: Lowest price at which a given stock traded during a regular trading session. ​

Close: Last price at which a given stock traded during a regular trading session. ​

Volume: Number of shares traded between the market open and close.​

Adj. Close: Adjusted stock close price after considering stock splits and dividends.​

Please also note that the New York Stock Exchange (NYSE) opens at 9:30 AM EST and the Shanghai Stock Exchange (SSE) opens at 9:30 AM China Standard Time.​

# CALCULATE PERCENTAGE DAILY RETURNS

```python
'''
Percentage Daily Returns
The stock percentage daily return rt indicates the percentage gain (or loss) for a given stock assuming that you bought the stock today and sold it the next day.
The percentage stock daily return is calculated using the following formula, where pt is the price of the stock at day t (today), p t-1 is the price of the stock at day t-1 (yesterday):
rt = (pt - p t-1) / p t-1
'''
```

In Python
Pandas.pct_change() method calculates the percentage change between current and prior elements in a Pandas DataFrame.

```python
import pandas as pd
import numpy as np
from google.colab import files
uploaded = files.upload()
```

```python
stock_df = pd.read_csv('Amazon.csv')
stock_df
```

```python
# Calculate the percentage daily return
stock_df['Daily Return'] = stock_df['Adj Close'].pct_change(1) * 100
stock_df
```

```python
# Let's replace the first row with zeros instead of NaN
stock_df['Daily Return'].replace(np.nan, 0, inplace = True)
stock_df
```

```python
# Use the describe() method to obtain a statistical summary about the data
# Over the specified time period, the average adjusted close price for Amazon stock was $120.07
# The maximum adjusted close price was $186.57
# The maximum volume of shares traded on one day were 311,346,000
stock_df.describe().round(2)
```
